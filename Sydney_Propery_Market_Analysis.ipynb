```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import OneHotEncoder


merged_data = pd.read_csv('domain_properties.csv')

# Data Preparation


# Feature Selection
# Select predictor variables (X) and target variable (y)
X = merged_data[['property_size', 'num_bed', 'num_bath']]
y = merged_data['price']

# One-Hot Encoding for Categorical Variable (suburb)
# Encoding the 'suburb' variable using one-hot encoding
encoder = OneHotEncoder(sparse=False, drop='first')
suburb_encoded = encoder.fit_transform(merged_data[['suburb']])
X_encoded = pd.concat([X, pd.DataFrame(suburb_encoded, columns=encoder.categories_[0][1:])], axis=1)


# Spliting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)


# Initializing and fit a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)


# Making predictions on the test set
y_pred = model.predict(X_test)

# Calculating evaluation metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared Score:", r2)


```
